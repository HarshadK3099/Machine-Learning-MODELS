Age,Estimated Salary are highly correlated with Purchased (response variable).

```python
x = df.iloc[:, [2, 3]].values  #age, estimatedSalary
y = df.iloc[:, 4].values     #Purchased
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
nb = GaussianNB()
nb.fit(X_train, Y_train)
```

```python
Y_pred = nb.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, Y_pred)
print('Confusion matrix:\n ',cm)
```

```python
print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
precision = TP / float(TP + FP)
recall = TP / float(TP + FN)
print(precision)
print(recall)
```

```python
print("Accuracy:", accuracy_score(Y_pred,Y_test))
print("Precision:", precision_score(Y_pred,Y_test))
print("Recall:", recall_score(Y_pred,Y_test))
print("F1 Score:", f1_score(Y_pred,Y_test))
print("\nClassification Report:\n", classification_report(Y_pred,Y_test))
```
