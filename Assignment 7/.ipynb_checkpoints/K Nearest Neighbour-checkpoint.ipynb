# Model Building

### Splitting the dataset

```python
X1 = df.drop(['User ID'],axis = 1)
X1
Y1 = df['User ID']
Y1
```

```python
#Extracting Independent and dependent Variable 
x = df.iloc[:, [2, 3]].values  #age, estimatedSalary
y = df.iloc[:, 4].values     #Purchased
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st= StandardScaler()    
X_train= st.fit_transform(X_train)    
X_test= st.transform(X_test)  
```

```python
print('Number of Data points in train data:',X_train.shape)
print('Number of Data points in test data:',Y_test.shape)
```

## Fitting of k-NN classifier to train data

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
KNeighborsClassifier?
```

```python
# Create a KNN classifier object
classifier = KNeighborsClassifier(n_neighbors=9, metric='minkowski', p=2 )  
classifier
```

```python
classifier.fit(X_train, Y_train)
```

### Predicting the Test Result

```python
y_pred = classifier.predict(X_test)
y_pred
```

### Evaluate the performance of the KNN classifier model

```python
# Evaluate the performance of the KNN classifier model
print('Accuracy:', accuracy_score(Y_test, y_pred))
print('Precision:', precision_score(Y_test, y_pred))
print('Recall:', recall_score(Y_test, y_pred))
print('F1-score:', f1_score(Y_test, y_pred))
```

### Creating the Confusion Matrix,Classification Report

```python
cm= confusion_matrix(Y_test, y_pred)  
print('Confusion Matrix:\n',cm)
print('\n')
print('Classification Report:\n',classification_report(Y_test,y_pred))
```

So from Above confusion matrix and classification report we can say that our model give good result on testing datasets with accuracy of 90%.

So KNN classifier is suitable for our datasets.
